```python
from collections import Counter
import pandas as pd
```

```python
def count_values(df, column):
    values = list(df[column])
    values = [e.split(' | ') for e in values]
    values = [e for sub in values for e in sub]
    return Counter(values)
```

```python
table_pointer = 'zrcsazu'
```

```python
df = pd.read_csv(f'gotriple_by_provider/csv/gotriple_response_{table_pointer}.csv')

counted_values = count_values(df, 'in_language')
in_language_counted = pd.DataFrame.from_dict(counted_values, orient='index', columns=['number_of_records']).sort_values(by='number_of_records', ascending=False)

counted_values = count_values(df, 'source')
source_counted = pd.DataFrame.from_dict(counted_values, orient='index', columns=['number_of_records']).sort_values(by='number_of_records', ascending=False)

with open(f'values_counting/gotriple_response_counted_{table_pointer}.csv', 'w', newline='', encoding='utf-8') as f:
    in_language_counted.to_csv(f)
    f.write('\n\n')
    source_counted.to_csv(f)
```
