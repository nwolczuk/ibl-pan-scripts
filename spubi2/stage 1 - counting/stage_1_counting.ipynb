```python
import json
```

```python
with open('extraction_output_stage_1.json', 'r', encoding='utf-8') as jfile:
    extraction_results = json.load(jfile)
```

```python
counting = {}

for key, value in extraction_results.items():
    for elem in value['nlp_results']:
        counting[elem['Etykieta']] = counting.setdefault(elem['Etykieta'], 0) + 1
    counting['all'] = counting.setdefault('all', 0) + 1
```
