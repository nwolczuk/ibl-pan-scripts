```python
import os
import json
```

```python
folder_path = 'anotacje_2_zbiorówki/output - etap 2/'
```

```python
file_names = [f for f in os.listdir(folder_path) if f.endswith('.json')]

counting = {}
for filename in file_names:
    with open(folder_path + filename, 'r', encoding='utf-8') as jfile:
        annotations_dict = json.load(jfile)
    if 'annotations' in annotations_dict:
        for elem in annotations_dict['annotations'][0][1]['entities']:
            counting[elem[2]] = counting.setdefault(elem[2], 0) + 1
```

```python
print(counting)
```
