```python
import pandas as pd
import json
```

```python
with open('validation_records.json', 'r', encoding='utf-8') as jfile:
    validation_records = json.load(jfile)
```

```python
to_df = []
for k, v in validation_records.items():
    nlp_results = [e for e in v['nlp_results']]
    nlp_results = [
        f"Oznaczony tekst: {e['Oznaczony tekst']}\nEtykieta: {e['Etykieta']}\nSkala pewności: {e['Skala pewności']}" 
        for e in nlp_results
    ]
    nlp_results = '\n\n'.join(nlp_results)
    to_df.append(
        (k, v['adnotacja'], nlp_results)
    )
```

```python
df = pd.DataFrame(to_df, columns=['ZAPIS_ID', 'ADNOTACJA', 'AUTOMATYCZNE_OZNACZENIA'])
df.to_excel('validation_records.xlsx', index=False)
```
