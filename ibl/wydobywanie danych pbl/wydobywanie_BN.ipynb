```python
import requests
import pandas as pd
```

```python
# autor
url = 'https://data.bn.org.pl/api/institutions/bibs.json?author=G%C3%B3ra%2C+Konrad+%281978-+%29&limit=100'
bibs = []

while True:
    print(url)
    resp = requests.get(url)
    if not resp.ok:
        break
    resp = resp.json()
    bibs.extend(resp.get('bibs'))
    if not resp.get('nextPage'):
        break
    else:
        url = resp.get('nextPage')

print(len(bibs))
```

```python
autor_records = []
for record in bibs:
    autor_records.append({
        "ID": record.get("id"),
        "Tytuł": record.get("title", ""),
        "Autor": record.get("author", ""),
        "ISBN/ISSN": record.get("isbnIssn", ""),
        "Wydawca": record.get("publisher", ""),
        "Miejsce wydania": record.get("placeOfPublication", ""),
        "Rok wydania": record.get("publicationYear", ""),
        'Temat': record.get("subject", ""),
        'Temat (dzieło)': record.get("subjectWork", ""),
        'Temat (czas)': record.get("subjectTime", ""),
        'Temat (miejsce)': record.get("subjectPlace", ""),
        "Język": record.get("language", ""),
        "Rodzaj materiału": record.get("kind", ""),
        "Numer bibliografii BN": record.get("nationalBibliographyNumber", "")
    })
```

```python
# temat
url = 'https://data.bn.org.pl/api/institutions/bibs.json?limit=100&subject=G%C3%B3ra%2C+Konrad+%281978-+%29'
bibs = []

while True:
    print(url)
    resp = requests.get(url)
    if not resp.ok:
        break
    resp = resp.json()
    bibs.extend(resp.get('bibs'))
    if not resp.get('nextPage'):
        break
    else:
        url = resp.get('nextPage')

print(len(bibs))
```

```python
temat_records = []
for record in bibs:
    temat_records.append({
        "ID": record.get("id"),
        "Tytuł": record.get("title", ""),
        "Autor": record.get("author", ""),
        "ISBN/ISSN": record.get("isbnIssn", ""),
        "Wydawca": record.get("publisher", ""),
        "Miejsce wydania": record.get("placeOfPublication", ""),
        "Rok wydania": record.get("publicationYear", ""),
        'Temat': record.get("subject", ""),
        'Temat (dzieło)': record.get("subjectWork", ""),
        'Temat (czas)': record.get("subjectTime", ""),
        'Temat (miejsce)': record.get("subjectPlace", ""),
        "Język": record.get("language", ""),
        "Rodzaj materiału": record.get("kind", ""),
        "Numer bibliografii BN": record.get("nationalBibliographyNumber", "")
    })
```

```python
df_aut = pd.DataFrame(autor_records)
df_sub = pd.DataFrame(temat_records)

with pd.ExcelWriter(f'BN_Góra_Konrad.xlsx', engine='openpyxl') as writer:
    df_aut.to_excel(writer, sheet_name='autor', index=False)
    df_sub.to_excel(writer, sheet_name='temat', index=False)
```
