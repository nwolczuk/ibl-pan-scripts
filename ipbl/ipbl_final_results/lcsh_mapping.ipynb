```python
import pandas as pd
```

```python
df = pd.read_excel('pbl_bn_mapping.xlsx')

def invert_chain(x):
    chain = x.split(' -> ')
    return ' -> '.join(chain[::-1])

df['inverted_chain'] = df['dzial'].apply(lambda x: invert_chain(x))

df.to_excel('pbl_bn_mapping_inverted_chains.xlsx')
```

```python
df = pd.read_excel('new_pbl_to_bn.xlsx', sheet_name='old_pbl_to_new_pbl')
df_lcsh = pd.read_excel('new_pbl_to_bn.xlsx', sheet_name='bn_to_lcsh')
df_lcsh = {row['desk BN']:row['link LCSH'] for _, row in df_lcsh.iterrows()}

df['new_pbl'] = df.new_pbl_id.astype(str) + ' ' + df.new_pbl_names.astype(str)
df = df[['new_pbl', 'bn_desc']]
df = df[df.new_pbl != 'nan nan']

def get_lcsh(x):
    bn_descs = str(x).split(' | ')
    lcsh_links = [df_lcsh.get(e) for e in bn_descs]
    lcsh_links = [e for e in lcsh_links if e]
    return ' | '.join(lcsh_links)

df['lcsh'] = df.bn_desc.apply(lambda x: get_lcsh(x))

df.to_excel('new_pbl_to_lcsh_mapping.xlsx')
```
