```python
import os
import json
from tqdm import tqdm
from ipbl_classifier import IPBLClassifier
```

```python
classifier = IPBLClassifier()
```

```python
list_of_files = os.listdir('automatic')
for filename in list_of_files:
    with open(f'automatic/{filename}', 'r', encoding='utf-8') as jfile:
        records = json.load(jfile)
    filtered_records = []
    all_results = []
    for record in tqdm(records):
        record_txt = record.get('Tekst artykułu')
        prediction = classifier.predict_pbl(record_txt)
        if prediction:
            filtered_records.append(record)
            all_results.append([record['Link'], True])
        else:
            all_results.append([record['Link'], False])
    print(filename, ' -- TRUE predicitions --> ', len(filtered_records))
    with open(f'filtered/filtered_{filename}', 'w', encoding='utf-8') as jfile:
        json.dump(filtered_records, jfile)
    with open(f'cassification/classification_{filename}', 'w', encoding='utf-8') as jfile:
        json.dump(all_results, jfile)
```
