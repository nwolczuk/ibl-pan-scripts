```python
import os
```

```python
list_of_files = os.listdir('bibframe_output/manual/')
counter = 0
for file in list_of_files:
    with open('bibframe_output/manual/'+file, 'r', encoding='utf-8') as txt:
        bibframe_str = ' '.join(txt.readlines())
        counter += bibframe_str.count('<bf:Work rdf:about="http://bibframe.example.org/works/')

print(counter)
```
